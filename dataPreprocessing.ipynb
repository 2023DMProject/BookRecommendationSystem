```python
import pandas as pd

books_df = pd.read_csv("Books.csv", dtype={"ISBN": str}, low_memory=False)
ratings_df = pd.read_csv("Ratings.csv", dtype={"ISBN": str}, low_memory=False)

filtered_rating_df = ratings_df[ratings_df["ISBN"].isin(books_df["ISBN"])]

filtered_rating_df.to_csv("Filtered_Ratings.csv", index=False)
```

```python
unique_isbn_cnt = filtered_rating_df["ISBN"].nunique()
unique_isbn_cnt
```

# 평점 분포 히스토그램

```python
import matplotlib.pyplot as plt
import seaborn as sns

#
plt.figure(figsize=(10, 6))
sns.histplot(filtered_rating_df['Book-Rating'], bins=15, kde=False)
plt.title('Book Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

# 가장 많은 평가를 받은 책 (상위 10개)

```python
top_books = filtered_rating_df['ISBN'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_books.plot(kind='bar')
plt.title('Top 10 Reviewed Books')
plt.xlabel('ISBN')
plt.ylabel('Number of Reviews')
plt.show()
```
