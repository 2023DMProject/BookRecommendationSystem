```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 파일 로드
# 예시 데이터 파일 형식: users.csv (유저 데이터), books.csv (책 데이터), ratings.csv (평점 데이터)
users = pd.read_csv('Users.csv')
books = pd.read_csv('Books.csv')
ratings = pd.read_csv('Filtered_Ratings.csv')

# 1. 평균 유저 나이 시각화
user_age_counts = users['Age'].value_counts().sort_index()
plt.bar(user_age_counts.index, user_age_counts.values)
plt.xlabel('age')
plt.ylabel('count')
plt.title('User Age Distribution')
plt.show()

# 2. 평균적으로 읽은 책의 수 시각화
books_per_user = ratings.groupby('User-ID')['ISBN'].size()
books_per_user_counts = books_per_user.value_counts().sort_index()
plt.hist(books_per_user, bins=30, edgecolor='black')
plt.xlabel('Book Count')
plt.ylabel('User Number')
plt.title('Read Book Distribution')
plt.xlim(0, 2000)
plt.show()


# 3. 책의 평균 평점(0점 제외) 시각화
average_ratings = ratings[ratings['Book-Rating'] > 0].groupby('ISBN')['Book-Rating'].mean()
average_ratings_counts = average_ratings.value_counts().sort_index()
plt.bar(average_ratings_counts.index, average_ratings_counts.values)
plt.xlabel('Rating')
plt.ylabel('Book Count')
plt.title('Book Rate Distriburion')
plt.show()
```
